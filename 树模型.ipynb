```python
import torch
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor


def target_fn(x):
    return np.sin(x) + 0.9*x+10

# 定义x的范围和间隔
x = np.linspace(-30, 30, 400)  # 从-10到10，取400个点
y = target_fn(x)


x = x[:, np.newaxis]
regressor = DecisionTreeRegressor(random_state=0)
regressor = regressor.fit(x,y)
# 预测
y_predict = regressor.predict(x)

plt.figure(figsize=(8,4))  # 设置图像大小
plt.plot(x, y, label="y =  0.9x + 10")
plt.plot(x, y_predict, label=f"Tree")

plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

print(f"loss is {mean_absolute_error(y_predict,y)}")
```

```python
# 生成未见过的数据, 超界
x_ = np.linspace(-60, 60, 700)  # 从-10到10，取400个点
y_ = target_fn(x_)

x_ = x_[:, np.newaxis]
y_predict_ = regressor.predict(x_)

plt.figure(figsize=(8,4))  # 设置图像大小
plt.plot(x_, y_, label="Trueth")
plt.plot(x_, y_predict_, label=f"Tree")

plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

print(f"loss is {mean_absolute_error(y_predict_,y_)}")
```

```python
# 生成未见过的数据, 密集
x_ = np.linspace(-30, 30, 1000)  # 从-10到10，取400个点
y_ = target_fn(x_)

x_ = x_[:, np.newaxis]
y_predict_ = regressor.predict(x_)

plt.figure(figsize=(8,4))  # 设置图像大小
plt.plot(x_, y_, label="Trueth")
plt.plot(x_, y_predict_, label=f"Tree")

plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

print(f"loss is {mean_absolute_error(y_predict_,y_)}")
```
